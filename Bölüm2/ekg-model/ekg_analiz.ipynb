> Accuracy , Precision , Recall , F1-score
* accuracy değeri diğer değerler ile uyumlu olmalı 
* diğer değerlerdn herhangi biri farklı olursa model değiştirilmeli 
* f1-score : Harmonik ortlama

gerçek   > dikey ise Tahmin 
[TP,FP]
[FN,TN]

Accuracy = (TP + TN )/ (TP+TN+FP+FN)
Precision = TP/(TP+FP)
Recall = TP/(TP+FN)
F1-score = 2 * Precision * Recall / (Precision + Recall)
