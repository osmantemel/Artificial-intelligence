```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

```

```python
data = pd.read_csv("sukalite.csv")
data.head()
```

```python
data = data.dropna()
data.isnull().sum()
```

```python

plt.figure(figsize=(15,10))
sns.countplot(data["Potability"])
plt.title("Distiribution of unsafe and safe water")

plt.show()
```
